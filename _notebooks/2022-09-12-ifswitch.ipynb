```python
import java.util.Date;
import java.text.ParseException;
import java.text.SimpleDateFormat;  
import java.util.Scanner;
public class ifswitch {
   

    static public void main(String[] args) throws ParseException { 
        Scanner check = new Scanner(System.in);
        System.out.println("Choose whether you'd like to call the if statement or the switch statement method:");
        String choice = check.nextLine();
        if (choice.equals("if") || choice.equals("i")) {
          long result = ageCheckerIf();
          System.out.println("You are " + result + " years old.");
        }
        else if (choice.equals("switch") || choice.equals("s")){
            long result = ageCheckerSwitch();
            System.out.println("You are" + result + "years old.");
        }
        else{
            System.out.println("Invalid input. Try again:");
            choice = check.nextLine();
        }
    check.close();
    }
public static long ageCheckerIf() throws ParseException{
    Scanner input = new Scanner(System.in);
    System.out.println("Enter a Birthday. (MM/DD/YYYY)");
    String birthday = input.nextLine();
    if (!birthday.matches(".*\\d{1}/\\d{1}/\\d{4}.*")) {
    
        System.out.println("Invalid Date");
        input.close();
        System.exit(0);
    }
    

    String[] birthdayArray = birthday.split("/");
    if(birthdayArray[0].length() != 2){
        birthdayArray[0] = "0" + birthdayArray[0];
    }
    if(birthdayArray[1].length() != 2){
        birthdayArray[1] = "0" + birthdayArray[1];
    }
    Date date1 = new SimpleDateFormat("MM/dd/yyyy").parse(birthday);  
    Date date = new Date();
    long difference = date.getTime() - date1.getTime();
    long differenceYears = (difference / (1000l*60*60*24*365));
    input.close();
    return differenceYears;



}

public static long ageCheckerSwitch() throws ParseException {
    Scanner input = new Scanner(System.in);
    System.out.println("Enter a Birthday. (MM/DD/YYYY)");
    String birthday = input.nextLine();
    if (!birthday.matches(".*\\d{1}/\\d{1}/\\d{4}.*")) {
    
        System.out.println("Invalid Date");
        input.close();
        System.exit(0);
    }
    

    String[] birthdayArray = birthday.split("/");
    switch(birthdayArray[0].length()){
        case 1:
            birthdayArray[0] = "0" + birthdayArray[0];
            break;
        case 2:
            break;
        default:
            System.out.println("Invalid Date");
            input.close();
            System.exit(0);
    }
    switch(birthdayArray[1].length()){
        case 1:
            birthdayArray[1] = "0" + birthdayArray[1];
            break;
        case 2:
            break;
        default:
            System.out.println("Invalid Date");
            input.close();
            System.exit(0);
    }
    
    Date date1 = new SimpleDateFormat("MM/dd/yyyy").parse(birthday);  
    Date date = new Date();
    long difference = date.getTime() - date1.getTime();
    long differenceYears = (difference / (1000l*60*60*24*365));
    input.close();
    return differenceYears;
}

}

```
